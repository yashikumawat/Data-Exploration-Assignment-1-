```python
# Step-3 Handling missing values
df.isnull().sum()
```

```python
df.drop(columns=['videos', 'what_customers_said'], inplace=True)
df
```

```python
df['discount'] = df['discount'].fillna(df['discount'].median())
```

```python
df['seller_name']=df['seller_name'].fillna("Unknown")
df['seller_information']=df['seller_information'].fillna("Not Available")
df['variations']=df['variations'].fillna("No Variations")
df
```

```python
df.isnull().sum()
```

```python
# Step-5 Removing duplicates
df.duplicated().sum()
```

```python
# Step-6 Create a derived column (savings_amount = price * discount/100)
df['savings_amount'] = df['initial_price'] * df['discount']/ 100
df
```

```python
# Step-7 Saving the cleaned dataset as a new CSV file
df.to_csv("Cleaned_dataset.csv", index=False)
```
